```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# ============================================================
# 1. Paths to your summary CSVs
# ============================================================

folder = "/project/biocomplexity/fa7sa/LLM_migration/results/LLM_to_LLM_summary/"  # change this

kendall_path = os.path.join(folder, "cross_llm_kendall_tau_summary.csv")
dtw_path     = os.path.join(folder, "cross_llm_dtw_summary.csv")
ttest_path   = os.path.join(folder, "cross_llm_ttest_summary.csv")

kendall_df = pd.read_csv(kendall_path)
dtw_df     = pd.read_csv(dtw_path)
ttest_df   = pd.read_csv(ttest_path)

# ------------------------------------------------------------
# Expected columns (based on what you shared):
#
# Kendall file:
#   Place, LLM_A, LLM_B,
#   Kendall_Tau_mean, Kendall_Tau_min, Kendall_Tau_max,
#   P_value_min, P_value_max,
#   High_Agreement_Count, Total_Comparisons
#
# DTW file:
#   Place, LLM_A, LLM_B,
#   DTW_Distance_mean, DTW_Distance_min, DTW_Distance_max, DTW_Distance_std,
#   Normalized_Distance_mean, Normalized_Distance_min, Normalized_Distance_max,
#   Normalized_Distance_std,
#   Very_Similar_Count, Similar_Count, Total_Comparisons
#
# T-test file (assumed):
#   Place, LLM_A, LLM_B,
#   T_stat_abs_mean, T_stat_abs_min, T_stat_abs_max,
#   P_value_mean, P_value_min, P_value_max,
#   Significant_Count, Total_Comparisons
#
# If your t-test columns differ, adjust below.
# ------------------------------------------------------------


# ============================================================
# 2. Helper to make grouped bar panel with error bars
# ============================================================

def plot_grouped_bar_with_minmax(
    ax,
    df,
    metric_mean,
    metric_min,
    metric_max,
    title,
    ylabel
):
    """
    df must contain:
      - Place
      - EITHER (LLM_A and LLM_B) OR precomputed LLM_pair
      - metric_mean, metric_min, metric_max

    It will:
      - ensure LLM_pair exists
      - group by Place and LLM_pair (implicitly via sorting)
      - plot mean with min/max error bars
    """
    df = df.copy()

    # If LLM_pair not present but LLM_A/B are, construct it
    if "LLM_pair" not in df.columns:
        if {"LLM_A", "LLM_B"}.issubset(df.columns):
            df["LLM_pair"] = df["LLM_A"] + "" + df["LLM_B"]
        else:
            raise ValueError(
                "DataFrame must have either LLM_pair or both LLM_A and LLM_B columns."
            )

    # Sort for consistent ordering
    df = df.sort_values(["LLM_pair", "Place"])

    llm_pairs = df["LLM_pair"].unique()
    places = df["Place"].unique()

    x = np.arange(len(llm_pairs))
    n_places = len(places)

    width = 0.6 / n_places if n_places > 0 else 0.3

    for i, place in enumerate(places):
        sub = df[df["Place"] == place].sort_values("LLM_pair")

        means = sub[metric_mean].values
        mins  = sub[metric_min].values
        maxs  = sub[metric_max].values

        lower_err = means - mins
        upper_err = maxs - means
        errors = [lower_err, upper_err]

        ax.bar(
            x + i * width,
            means,
            width=width,
            yerr=errors,
            capsize=5,
            alpha=0.85,
            label=place,
        )

    ax.set_title(title, fontsize=12)
    ax.set_ylabel(ylabel)
    ax.set_xticks(x + width * (n_places - 1) / 2.0)
    ax.set_xticklabels(llm_pairs, rotation=20, ha="right")
    ax.legend(frameon=False)


# ============================================================
# 3. Prepare any derived columns
# ============================================================

# For t-test, define a "fraction of significant differences" if columns exist
if {"Significant_Count", "Total_Comparisons"}.issubset(ttest_df.columns):
    ttest_df = ttest_df.copy()
    ttest_df["Significant_Fraction"] = (
        ttest_df["Significant_Count"] / ttest_df["Total_Comparisons"].replace(0, np.nan)
    )
else:
    # If you do not have Significant_Count/Total_Comparisons, you can instead
    # base the panel on |t| mean. Adjust the plotting call below accordingly.
    pass


# ============================================================
# 4. Build the 3-panel figure
# ============================================================

fig, axes = plt.subplots(3, 1, figsize=(10, 13), constrained_layout=True)

# ---------------- Panel 1: Kendall tau ----------------------
plot_grouped_bar_with_minmax(
    ax=axes[0],
    df=kendall_df,
    metric_mean="Kendall_Tau_mean",
    metric_min="Kendall_Tau_min",
    metric_max="Kendall_Tau_max",
    title="Trend Agreement Between LLMs (Kendall )",
    ylabel=" (mean  range)"
)

# ---------------- Panel 2: Normalized DTW -------------------
plot_grouped_bar_with_minmax(
    ax=axes[1],
    df=dtw_df,
    metric_mean="Normalized_Distance_mean",
    metric_min="Normalized_Distance_min",
    metric_max="Normalized_Distance_max",
    title="Shape Distance Between LLMs (Normalized DTW)",
    ylabel="Normalized DTW (lower = more similar)"
)

# ---------------- Panel 3: t-test differences ---------------
# Option 1: plot fraction of significant differences (recommended)
if "Significant_Fraction" in ttest_df.columns:
    # Here we do not use min/max, because fraction is already an aggregate.
    # Instead, we plot mean fraction across runs per PlaceLLM_pair.
    df_t = ttest_df.copy()
    df_t["LLM_pair"] = df_t["LLM_A"] + "" + df_t["LLM_B"]
    grouped = df_t.groupby(["Place", "LLM_pair"], as_index=False)["Significant_Fraction"].mean()

    tmp = grouped.rename(columns={"Significant_Fraction": "Fraction_mean"})
    # Create fake min/max columns equal to mean so we can reuse the helper
    tmp["Fraction_min"] = tmp["Fraction_mean"]
    tmp["Fraction_max"] = tmp["Fraction_mean"]

    plot_grouped_bar_with_minmax(
        ax=axes[2],
        df=tmp,
        metric_mean="Fraction_mean",
        metric_min="Fraction_min",
        metric_max="Fraction_max",
        title="Magnitude Difference Between LLMs (t-test)",
        ylabel="Fraction of runs with p < 0.05"
    )
else:
    # Option 2: fallback to |t-stat| mean with min/max
    plot_grouped_bar_with_minmax(
        ax=axes[2],
        df=ttest_df,
        metric_mean="T_stat_abs_mean",
        metric_min="T_stat_abs_min",
        metric_max="T_stat_abs_max",
        title="Magnitude Difference Between LLMs (t-test)",
        ylabel="|t-stat| (mean  range)"
    )

fig.suptitle("Pairwise LLM Agreement Across Trend, Shape, and Magnitude", fontsize=16)

# Save to file for poster use
out_path = os.path.join(folder, "panel_A_llm_to_llm_summary.png")
fig.savefig(out_path, dpi=300, bbox_inches="tight")

plt.show()

print(f"Saved figure to: {out_path}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Load your GT summary file
# -----------------------------
gt_path = "/project/biocomplexity/fa7sa/LLM_migration/results/ground_truth_comparison/all_places_ground_truth_summary.csv"
df = pd.read_csv(gt_path)

# Expected columns (adjust if needed):
# Place, LLM, Kendall_Tau_mean, Normalized_Distance_mean,
# Significant_Fraction or p-values

# If the file uses LLM_A, rename:
if "LLM_A" in df.columns:
    df = df.rename(columns={"LLM_A": "LLM"})

# -----------------------------
# Extract countries and LLMs
# -----------------------------
places = df["Place"].unique()
llms = df["LLM"].unique()

# -----------------------------
# Prepare metrics for radar
# -----------------------------
# Normalize values so that 1 = best, 0 = worst

def normalize_inverse(x):
    """Convert 'lower is better' metric to 'higher is better' on [0,1]."""
    # If constant, avoid division by zero
    if x.max() == x.min():
        return np.ones_like(x)
    return 1 - (x - x.min()) / (x.max() - x.min())

# For each country build a dict: {LLM: [, shape_score, magnitude_score]}
results = {place: {} for place in places}

for place in places:
    sub = df[df["Place"] == place].copy().reset_index(drop=True)

    # Trend = Kendall  (already higher=better)
    tau = sub["Kendall_Tau"].values

    # Shape = inverse-normalized DTW
    dtw = sub["Normalized_Distance"].values
    dtw_score = normalize_inverse(dtw)

    # Magnitude = inverse of significant fraction OR use p-value
    if "Significant_Fraction" in sub.columns:
        mag = sub["Significant_Fraction"].values
        mag_score = 1 - mag  # lower significance fraction = better
    elif "P_value_mean" in sub.columns:
        mag = sub["P_value_mean"].values
        mag_score = mag / mag.max()  # scale so higher p-value = better
    else:
        raise KeyError("No magnitude column found (Significant_Fraction or P_value_mean).")

    for i, llm in enumerate(sub["LLM"]):
        results[place][llm] = [
            tau[i],         # Trend
            dtw_score[i],   # Shape
            mag_score[i],   # Magnitude
        ]

# -----------------------------
# Radar plot helper
# -----------------------------
def make_radar(ax, data_dict, title):
    labels = ["Trend ()", "Shape Similarity", "Magnitude Match"]
    num_vars = len(labels)

    # Angles for radar
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False)
    angles = np.concatenate((angles, [angles[0]]))

    for llm, values in data_dict.items():
        vals = np.array(values)
        vals = np.concatenate((vals, [vals[0]]))
        ax.plot(angles, vals, label=llm, marker="o", linewidth=2)
        ax.fill(angles, vals, alpha=0.15)

    ax.set_title(title, fontsize=14)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)
    ax.set_yticklabels([])
    ax.set_ylim(0, 1)
    ax.legend(loc="upper right", bbox_to_anchor=(1.2, 1.1))

# -----------------------------
# Create figure with 2 radars
# -----------------------------
fig, axes = plt.subplots(1, 2, subplot_kw=dict(polar=True), figsize=(14, 7))

for ax, place in zip(axes, places):
    make_radar(ax, results[place], f"LLM vs Ground Truth  {place}")

plt.suptitle("LLM Forecast Accuracy Compared to Ground Truth", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ============================
# Load data
# ============================
csv_path = "/project/biocomplexity/fa7sa/LLM_migration/results/ground_truth_comparison/all_places_ground_truth_summary.csv"  # change if needed
df = pd.read_csv(csv_path)

# Expected columns:
# Place, LLM, Run, Used_Points, Percent_Used,
# t_stat, TTest_P_value,
# Kendall_Tau, Tau_P_value, Tau_Strength,
# DTW_Distance, Normalized_Distance, DTW_Interpretation

# ============================
# Aggregate over runs
# ============================
agg = (
    df.groupby(["Place", "LLM"], as_index=False)
      .agg({
          "Kendall_Tau": "mean",
          "Normalized_Distance": "mean",
          "TTest_P_value": "mean"
      })
)

# ============================
# Build scores in [0, 1]
#   Trend: map  from [-1,1]  [0,1]
#   Shape: invert-normalize Normalized_Distance (lower is better)
#   Magnitude: normalize p-value (higher is better)
# ============================

# Trend score
agg["Trend_score"] = (agg["Kendall_Tau"] + 1.0) / 2.0
agg["Trend_score"] = agg["Trend_score"].clip(0.0, 1.0)

# Shape score (inverse-normalized DTW)
nd = agg["Normalized_Distance"].values
if nd.max() == nd.min():
    agg["Shape_score"] = 1.0
else:
    agg["Shape_score"] = 1.0 - (nd - nd.min()) / (nd.max() - nd.min())

# Magnitude score (p-value, normalized)
pv = agg["TTest_P_value"].values
if pv.max() == pv.min():
    agg["Magnitude_score"] = 1.0
else:
    agg["Magnitude_score"] = (pv - pv.min()) / (pv.max() - pv.min())

# ============================
# Prepare data dict for radars
# results[place][llm] = [trend, shape, magnitude]
# ============================
results = {}
for place in agg["Place"].unique():
    sub = agg[agg["Place"] == place]
    results[place] = {}
    for _, row in sub.iterrows():
        llm = row["LLM"]
        results[place][llm] = [
            row["Trend_score"],
            row["Shape_score"],
            row["Magnitude_score"],
        ]

# ============================
# Radar plot helper
# ============================
def make_radar(ax, data_dict, title):
    labels = ["Trend ()", "Shape similarity", "Magnitude match"]
    num_vars = len(labels)

    # Angles for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False)
    angles = np.concatenate([angles, [angles[0]]])  # close the loop

    for llm, vals in data_dict.items():
        vals = np.array(vals, dtype=float)
        vals = np.concatenate([vals, [vals[0]]])  # close the loop

        ax.plot(angles, vals, marker="o", linewidth=2, label=llm)
        ax.fill(angles, vals, alpha=0.15)

    ax.set_title(title, fontsize=14)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)
    ax.set_yticklabels([])
    ax.set_ylim(0.0, 1.0)
    ax.grid(True)
    ax.legend(loc="upper right", bbox_to_anchor=(1.25, 1.05), frameon=False)

# ============================
# Create one radar per Place
# ============================
places = list(results.keys())
n_places = len(places)

fig, axes = plt.subplots(
    1, n_places,
    subplot_kw=dict(polar=True),
    figsize=(6 * n_places, 6),
)

# Handle case of single place (axes not an array)
if n_places == 1:
    axes = [axes]

for ax, place in zip(axes, places):
    make_radar(ax, results[place], f"LLM vs Ground Truth  {place}")

fig.suptitle("LLM Forecast Agreement with Ground Truth\nTrend, Shape, and Magnitude", fontsize=16)
plt.tight_layout()
plt.savefig("radar_chart.pdf")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ============================
# 1. Load data
# ============================
csv_path = "/project/biocomplexity/fa7sa/LLM_migration/results/ground_truth_comparison/all_places_ground_truth_summary.csv" 
df = pd.read_csv(csv_path)

# Expected columns:
# Place, LLM, Run, Used_Points, Percent_Used,
# t_stat, TTest_P_value,
# Kendall_Tau, Tau_P_value, Tau_Strength,
# DTW_Distance, Normalized_Distance, DTW_Interpretation

# ============================
# 2. Build a meaningful magnitude score:
#    Magnitude_score = fraction of runs where p > 0.05
#    (i.e., NOT significantly different from GT)
# ============================
df = df.copy()
df["Not_Significant"] = (df["TTest_P_value"] > 0.05).astype(float)

agg = (
    df.groupby(["Place", "LLM"], as_index=False)
      .agg(
          Kendall_Tau_mean=("Kendall_Tau", "mean"),
          Normalized_Distance_mean=("Normalized_Distance", "mean"),
          NonSig_fraction=("Not_Significant", "mean"),
      )
)

# ============================
# 3. Convert metrics to radar scores in [0,1]
#    but keep original maxima for labeling
# ============================
def compute_scores_for_place(sub_df):
    """
    sub_df: rows for a single Place with columns:
      LLM, Kendall_Tau_mean, Normalized_Distance_mean, NonSig_fraction
    Returns:
      scores: dict[llm] -> [trend_score, shape_score, magnitude_score]
      axis_max: dict with max original values for labeling
    """
    # Original max values (for annotation, not normalization)
    tau_max_orig = sub_df["Kendall_Tau_mean"].max()
    dtw_max_orig = sub_df["Normalized_Distance_mean"].max()
    frac_max_orig = sub_df["NonSig_fraction"].max()

    axis_max = {
        "tau_max": tau_max_orig,
        "dtw_max": dtw_max_orig,
        "frac_max": frac_max_orig,
    }

    scores = {}

    for _, row in sub_df.iterrows():
        llm = row["LLM"]

        # Trend score: map  from [-1,1] to [0,1]
        tau = row["Kendall_Tau_mean"]
        trend_score = (tau + 1.0) / 2.0
        trend_score = float(np.clip(trend_score, 0.0, 1.0))

        # Shape score: your choice A  1 / (1 + Normalized_Distance)
        nd = row["Normalized_Distance_mean"]
        shape_score = 1.0 / (1.0 + nd)

        # Magnitude score: fraction of runs with p > 0.05 (already 01)
        mag_score = float(np.clip(row["NonSig_fraction"], 0.0, 1.0))

        scores[llm] = [trend_score, shape_score, mag_score]

    return scores, axis_max


# ============================
# 4. Radar plot helper
# ============================
def make_radar(ax, scores_dict, axis_max, title):
    """
    scores_dict: { LLM: [trend_score, shape_score, magnitude_score] }
    axis_max: dict with original max values:
        tau_max, dtw_max, frac_max
    """
    labels = ["Trend (Kendall )", "Shape similarity \n (Normalize Inverse DTW)", "Magnitude match (p>0.05)"]
    num_vars = len(labels)

    # Angles for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False)
    angles = np.concatenate([angles, [angles[0]]])  # close the loop

    # Plot each LLM polygon
    for llm, vals in scores_dict.items():
        vals = np.array(vals, dtype=float)
        vals = np.concatenate([vals, [vals[0]]])  # close the loop

        ax.plot(angles, vals, marker="o", linewidth=2, label=llm)
        ax.fill(angles, vals, alpha=0.15)

    # Basic radar styling
    ax.set_title(title, fontsize=14)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)
    ax.set_yticklabels([])
    ax.set_ylim(-1.0, 1.0)
    ax.grid(True)
    ax.legend(loc="upper right", bbox_to_anchor=(1.25, 1.05), frameon=False)

    # ============================
    # Label each axis end with the corresponding max *original* value
    # ============================
    tau_max = axis_max["tau_max"]
    dtw_max = axis_max["dtw_max"]
    frac_max = axis_max["frac_max"]

    max_labels = [
        f"max  = {tau_max:.2f}",
        f"max DTW = {dtw_max:.1f}",
        f"max frac = {frac_max:.2f}",
    ]

    # for angle, text in zip(angles[:-1], max_labels):
    #     ax.text(
    #         angle,
    #         1.05,  # just outside the unit circle
    #         text,
    #         ha="center",
    #         va="center",
    #         fontsize=9,
    #     )


# ============================
# 5. Make one radar per Place
# ============================
places = agg["Place"].unique()

fig, axes = plt.subplots(
    1, len(places),
    subplot_kw=dict(polar=True),
    figsize=(6 * len(places), 6),
)

if len(places) == 1:
    axes = [axes]

for ax, place in zip(axes, places):
    sub = agg[agg["Place"] == place].reset_index(drop=True)
    scores_dict, axis_max = compute_scores_for_place(sub)
    make_radar(ax, scores_dict, axis_max, f"LLM vs Ground Truth  {place}")

fig.suptitle(
    "LLM Forecast Agreement with Ground Truth\n"
    "Trend (), Shape (DTW-based), and Magnitude (fraction non-significant)",
    fontsize=16,
)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load data
romania_df = pd.read_csv('/project/biocomplexity/fa7sa/LLM_migration/results/ground_truth_comparison/Romania_ground_truth_combined_summary.csv')
slovakia_df = pd.read_csv('/project/biocomplexity/fa7sa/LLM_migration/results/ground_truth_comparison/Slovakia_ground_truth_combined_summary.csv')
# Combine both places
combined_df = pd.concat([romania_df, slovakia_df], ignore_index=True)

# DEBUG: Check what the actual LLM names are in your data
print("Unique LLM names in data:", combined_df['LLM'].unique())
print("Unique Place names in data:", combined_df['Place'].unique())
print("\nSample of data:")
print(combined_df[['Place', 'LLM', 'Kendall_Tau', 'TTest_P_value']].head(10))

# Standardize the LLM names (convert to lowercase to match)
combined_df['LLM'] = combined_df['LLM'].str.lower().str.strip()
combined_df['Place'] = combined_df['Place'].str.lower().str.strip()

print("\nAfter standardization:")
print("Unique LLM names:", combined_df['LLM'].unique())
print("Unique Place names:", combined_df['Place'].unique())

def create_ground_truth_heatmap():
    """
    Create a compact heatmap showing both Kendall's Tau and T-test results.
    """
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Get actual LLM names and places from the data
    llm_order = sorted(combined_df['LLM'].unique())
    places = sorted(combined_df['Place'].unique())
    
    print(f"\nCreating heatmap for LLMs: {llm_order}")
    print(f"Places: {places}")
    
    # --- PANEL A: Kendall's Tau Heatmap ---
    tau_matrix = np.zeros((len(llm_order), len(places)))
    
    for i, llm in enumerate(llm_order):
        for j, place in enumerate(places):
            data = combined_df[(combined_df['LLM'] == llm) & (combined_df['Place'] == place)]
            if len(data) > 0:
                tau_matrix[i, j] = data['Kendall_Tau'].mean()
                print(f"{llm} - {place}: Kendall's Tau = {tau_matrix[i, j]:.3f} (n={len(data)})")
            else:
                tau_matrix[i, j] = np.nan
                print(f"{llm} - {place}: NO DATA FOUND")
    
    im1 = axes[0].imshow(tau_matrix, cmap='RdYlGn', aspect='auto', vmin=-1, vmax=1)
    
    # Add text annotations
    for i in range(len(llm_order)):
        for j in range(len(places)):
            value = tau_matrix[i, j]
            if not np.isnan(value):
                # Choose text color based on background
                text_color = 'white' if abs(value) > 0.5 else 'black'
                text = axes[0].text(j, i, f'{value:.2f}',
                                   ha="center", va="center", color=text_color, 
                                   fontsize=14, fontweight='bold')
            else:
                text = axes[0].text(j, i, 'N/A',
                                   ha="center", va="center", color="gray", 
                                   fontsize=12, style='italic')
    
    axes[0].set_xticks(np.arange(len(places)))
    axes[0].set_yticks(np.arange(len(llm_order)))
    axes[0].set_xticklabels([p.capitalize() for p in places], fontsize=12)
    axes[0].set_yticklabels([llm.capitalize() for llm in llm_order], fontsize=12)
    axes[0].set_title("A) Trend Agreement (Kendall's )\nwith Ground Truth", 
                      fontsize=13, fontweight='bold', pad=10)
    
    # Colorbar
    cbar1 = plt.colorbar(im1, ax=axes[0], fraction=0.046, pad=0.04)
    cbar1.set_label("Kendall's ", fontsize=11, fontweight='bold')
    
    # --- PANEL B: Magnitude Similarity Heatmap ---
    similarity_matrix = np.zeros((len(llm_order), len(places)))
    
    for i, llm in enumerate(llm_order):
        for j, place in enumerate(places):
            data = combined_df[(combined_df['LLM'] == llm) & (combined_df['Place'] == place)]
            if len(data) > 0:
                # Fraction of runs that are NOT significantly different from GT
                similarity_matrix[i, j] = (data['TTest_P_value'] >= 0.05).mean()
                print(f"{llm} - {place}: Similarity = {similarity_matrix[i, j]*100:.1f}%")
            else:
                similarity_matrix[i, j] = np.nan
    
    im2 = axes[1].imshow(similarity_matrix, cmap='RdYlGn', aspect='auto', vmin=0, vmax=1)
    
    # Add text annotations with percentages
    for i in range(len(llm_order)):
        for j in range(len(places)):
            value = similarity_matrix[i, j]
            if not np.isnan(value):
                # Choose text color based on background
                text_color = 'white' if value < 0.3 or value > 0.7 else 'black'
                text = axes[1].text(j, i, f'{value*100:.0f}%',
                                   ha="center", va="center", color=text_color, 
                                   fontsize=14, fontweight='bold')
            else:
                text = axes[1].text(j, i, 'N/A',
                                   ha="center", va="center", color="gray", 
                                   fontsize=12, style='italic')
    
    axes[1].set_xticks(np.arange(len(places)))
    axes[1].set_yticks(np.arange(len(llm_order)))
    axes[1].set_xticklabels([p.capitalize() for p in places], fontsize=12)
    axes[1].set_yticklabels([llm.capitalize() for llm in llm_order], fontsize=12)
    axes[1].set_title('B) Magnitude Similarity\n(% Runs with p  0.05)', 
                      fontsize=13, fontweight='bold', pad=10)
    
    # Colorbar
    cbar2 = plt.colorbar(im2, ax=axes[1], fraction=0.046, pad=0.04)
    cbar2.set_label('Fraction Similar', fontsize=11, fontweight='bold')
    
    plt.suptitle('LLM Accuracy vs Ground Truth', fontsize=16, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.savefig('ground_truth_heatmap_poster.png', dpi=300, bbox_inches='tight')
    plt.savefig('ground_truth_heatmap_poster.pdf', bbox_inches='tight')
    plt.show()
    
    # Print summary statistics
    print("\n" + "="*70)
    print("SUMMARY: Kendall's Tau with Ground Truth")
    print("="*70)
    print("\nTau Matrix (rows=LLMs, cols=places):")
    print(pd.DataFrame(tau_matrix, index=[llm.capitalize() for llm in llm_order], 
                      columns=[p.capitalize() for p in places]))
    
    print("\n" + "="*70)
    print("SUMMARY: Magnitude Similarity (% runs NOT significantly different)")
    print("="*70)
    print("\nSimilarity Matrix (rows=LLMs, cols=places):")
    print(pd.DataFrame(similarity_matrix * 100, index=[llm.capitalize() for llm in llm_order], 
                      columns=[p.capitalize() for p in places]))

# Run the visualization
create_ground_truth_heatmap()
```

```python
import plotly.graph_objects as go

# ---------------------------------------
# 1. Define nodes
# ---------------------------------------

# Level 2 nodes: encode Level 1 + Level 2 in the label
level2_labels = [
    "Linear  Simple",
    "Linear  Autoregressive",
    "Linear  Multi-phase",
    "Non-Linear  Simple",
    "Non-Linear  Multi-phase"
]

model_labels = ["GPT-5", "Sonnet-4.5", "Gemini-2.5"]

labels = level2_labels + model_labels

# Node indices
# 0..4  -> Level 2 nodes
# 5..7  -> Model nodes
idx = {label: i for i, label in enumerate(labels)}

# ---------------------------------------
# 2. Define your percentages (example values)
#    Replace these with your actual % (or counts).
#    Values do not have to be normalized; Sankey will scale them.
# ---------------------------------------

# Example structure: (level2_label, model_label): value
flows = {
    ("Linear  Simple",         "GPT-5"):       30,
    ("Linear  Simple",         "Sonnet-4.5"):  40,
    ("Linear  Simple",         "Gemini-2.5"):  30,

    ("Linear  Autoregressive", "GPT-5"):       50,
    ("Linear  Autoregressive", "Sonnet-4.5"):  20,
    ("Linear  Autoregressive", "Gemini-2.5"):  30,

    ("Linear  Multi-phase",    "GPT-5"):       40,
    ("Linear  Multi-phase",    "Sonnet-4.5"):  35,
    ("Linear  Multi-phase",    "Gemini-2.5"):  25,

    ("Non-Linear  Simple",     "GPT-5"):       25,
    ("Non-Linear  Simple",     "Sonnet-4.5"):  45,
    ("Non-Linear  Simple",     "Gemini-2.5"):  30,

    ("Non-Linear  Multi-phase","GPT-5"):       35,
    ("Non-Linear  Multi-phase","Sonnet-4.5"):  35,
    ("Non-Linear  Multi-phase","Gemini-2.5"):  30,
}

# ---------------------------------------
# 3. Build source, target, value arrays
# ---------------------------------------

sources = []
targets = []
values = []

for (src_label, tgt_label), v in flows.items():
    sources.append(idx[src_label])
    targets.append(idx[tgt_label])
    values.append(v)

# ---------------------------------------
# 4. Optional: fix node positions for a vertical layout
#    Top row = Level 2 nodes, bottom row = models
# ---------------------------------------

# All Level 2 nodes at y = 0.1 (top), models at y = 0.8 (bottom)
node_x = [0.0, 0.25, 0.5, 0.75, 1.0] + [0.2, 0.5, 0.8]
node_y = [0.1] * 5 + [0.8] * 3

# Colors per model (kept consistent with your other plots)
model_colors = {
    "GPT-5":       "rgba(31, 119, 180, 0.8)",   # blue
    "Sonnet-4.5":  "rgba(255, 127, 14, 0.8)",   # orange
    "Gemini-2.5":  "rgba(44, 160, 44, 0.8)",    # green
}

# Node colors: two palettes for Linear vs Non-Linear to keep tree feeling
node_colors = []
for lab in labels:
    if lab.startswith("Linear"):
        node_colors.append("rgba(200, 220, 255, 1.0)")      # light blue
    elif lab.startswith("Non-Linear"):
        node_colors.append("rgba(255, 220, 200, 1.0)")      # light orange
    else:
        # Model nodes: use their main color
        node_colors.append(model_colors.get(lab, "rgba(200,200,200,1.0)"))

# Link colors: inherit model color with transparency
link_colors = []
for (src_label, tgt_label), v in flows.items():
    link_colors.append(model_colors.get(tgt_label, "rgba(150,150,150,0.4)").replace("1.0", "0.4"))

# ---------------------------------------
# 5. Create vertical Sankey
# ---------------------------------------

sankey = go.Sankey(
    orientation="v",  # vertical flow: top -> bottom
    arrangement="snap",
    node=dict(
        pad=20,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels,
        color=node_colors,
        x=node_x,
        y=node_y
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
        color=link_colors
    )
)

fig = go.Figure(data=[sankey])

fig.update_layout(
    title=dict(
        text="Hierarchy of Analytical Models and LLM Usage",
        x=0.5,
        xanchor="center",
        y=0.95
    ),
    font=dict(size=14),
    margin=dict(l=20, r=20, t=60, b=20)
)

fig.show()
# For saving static figure (if you have kaleido installed):
# fig.write_image("model_hierarchy_vertical_sankey.png", scale=3)
fig.write_image("model_hierarchy_vertical_sankey.pdf", scale=3)
```
